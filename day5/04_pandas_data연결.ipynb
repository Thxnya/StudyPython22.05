```python
person = pd.read_csv('C:/DEV/STUDY/StudyPython22.05/data/survey_person.csv')
site = pd.read_csv('C:/DEV/STUDY/StudyPython22.05/data/survey_site.csv')
survey = pd.read_csv('C:/DEV/STUDY/StudyPython22.05/data/survey_survey.csv')
visited = pd.read_csv('C:/DEV/STUDY/StudyPython22.05/data/survey_visited.csv')
```

```python
sv_merge = site.merge(visited, left_on='name', right_on='site')
print(sv_merge)
```

```python
ps_merge = person.merge(survey, left_on='ident', right_on='person')
print(ps_merge)
# survey의 person열에서 NaN값은 person의 ident열과 매칭이 안되기때문에 빠짐
```

```python
vs_merge = visited.merge(survey, left_on='ident', right_on='taken')
print(vs_merge)
```

```python
sp = survey.merge(person, left_on='person', right_on='ident')
print(sp)
```

```python
spv = sp.merge(visited, left_on='taken', right_on = 'ident')
print(spv)
```

```python
spvs = spv.merge(site, left_on='site', right_on='name')
print(spvs)
```
