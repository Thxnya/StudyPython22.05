```python
# 데이터 개수 확인
print(ebola.count())
```

```python
# 누락값 개수 확인
num_rows = ebola.shape[0]
num_missing = num_rows - ebola.count()
print(num_missing)
```

```python
import numpy as np
print(np.count_nonzero(ebola.isnull()))
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
```

```python
print(ebola['Cases_Guinea'].value_counts(dropna=False))
```

```python
ebola.iloc[:, 0:5].head()
```

```python
# NaN 값을 0으로 채움
ebola['Cases_Guinea'].fillna(0)
```

```python
print(ebola['Cases_Guinea'].mean())
print(ebola['Cases_Guinea'].fillna(0).mean())
```

```python
# 누락값이 나타나기 전 값으로 변경
ebola.fillna(method='ffill').iloc[0:5, 0:5]
```

```python
# 누락값이 나타난 이후 값으로 변경
ebola.fillna(method='bfill').iloc[0:5, 0:5]
```

```python
print(ebola['Cases_Guinea'].mean())
print(ebola['Cases_Guinea'].fillna(method='ffill').mean())
print(ebola['Cases_Guinea'].fillna(method='bfill').mean())
```

```python
# 누락값 양쪽에 있는 값을 이용하여 중간값을 구한 다음 변경
ebola.interpolate().iloc[0:5, 0:5]
```

```python
print(ebola.shape)
ebola_dropna = ebola.dropna()
print(ebola_dropna.shape)
print(ebola_dropna.iloc[0:5, 0:5])
```
